
# Probability Distributions: Normal, Binomial, Poisson, Exponential, and Other Key Distributions

## Overview
Probability distributions describe how values are distributed across different outcomes in a dataset or a random process. Each distribution has unique properties and applications in statistics, finance, engineering, and the natural sciences.

In this notebook, we will cover key distributions including:

- **Normal Distribution**: Symmetrical, bell-shaped distribution, commonly used in statistics.
- **Binomial Distribution**: Describes the number of successes in a fixed number of independent binary trials.
- **Poisson Distribution**: Models the number of events occurring in a fixed interval of time or space.
- **Exponential Distribution**: Models the time between events in a Poisson process.



## 1. Normal Distribution

The normal distribution is a continuous, symmetrical, bell-shaped distribution. It is characterized by its mean (\( \mu \)) and standard deviation (\( \sigma \)). The probability density function (PDF) for a normal distribution is:

\[ f(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{-\frac{(x - \mu)^2}{2 \sigma^2}} \]

Let's generate a normal distribution and visualize it.


```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0    # Mean
sigma = 1  # Standard deviation

# Generate data
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, mu, sigma)

# Plot
plt.plot(x, pdf, label="Normal Distribution (mu=0, sigma=1)")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.title("Normal Distribution")
plt.legend()
plt.show()
```


## 2. Binomial Distribution

The binomial distribution models the number of successes in a fixed number of independent trials, where each trial has a success probability \( p \). The probability mass function (PMF) for the binomial distribution is:

\[ P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k} \]

where \( n \) is the number of trials, \( k \) is the number of successes, and \( p \) is the probability of success.

Let's visualize a binomial distribution.


```python

from scipy.stats import binom

# Parameters for the binomial distribution
n = 20   # Number of trials
p = 0.5  # Probability of success

# Generate data
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

# Plot
plt.stem(x, pmf, basefmt=" ")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title("Binomial Distribution (n=20, p=0.5)")
plt.show()
```


## 3. Poisson Distribution

The Poisson distribution models the number of events that occur in a fixed interval of time or space, given a known average rate (\( \lambda \)). The probability mass function (PMF) for the Poisson distribution is:

\[ P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!} \]

where \( k \) is the number of events.

Let's visualize a Poisson distribution.


```python

from scipy.stats import poisson

# Parameter for the Poisson distribution
lambda_ = 3  # Average rate

# Generate data
x = np.arange(0, 15)
pmf = poisson.pmf(x, lambda_)

# Plot
plt.stem(x, pmf, basefmt=" ")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.title("Poisson Distribution (lambda=3)")
plt.show()
```


## 4. Exponential Distribution

The exponential distribution models the time between events in a Poisson process. The probability density function (PDF) for the exponential distribution is:

\[ f(x; \lambda) = \lambda e^{-\lambda x} \]

where \( \lambda \) is the rate parameter.

Let's visualize an exponential distribution.


```python

from scipy.stats import expon

# Parameter for the exponential distribution
lambda_ = 1  # Rate parameter

# Generate data
x = np.linspace(0, 5, 1000)
pdf = expon.pdf(x, scale=1/lambda_)

# Plot
plt.plot(x, pdf, label="Exponential Distribution (lambda=1)")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.title("Exponential Distribution")
plt.legend()
plt.show()
```
