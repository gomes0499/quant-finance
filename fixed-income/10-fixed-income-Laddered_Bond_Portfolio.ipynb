
# Laddered Bond Portfolio: Structuring a Portfolio for Regular Maturities

## Overview
A laddered bond portfolio is a strategy that involves structuring a bond portfolio with bonds that mature at regular intervals. This approach helps manage liquidity needs and reduces interest rate risk, as the portfolio will consistently have bonds maturing and reinvested at prevailing rates.

In this notebook, we will:

- **Explore the Laddering Strategy**: Understand how a laddered bond portfolio helps manage liquidity and interest rate risk.
- **Simulate a Laddered Portfolio**: Structure a hypothetical portfolio with bonds maturing annually over a 10-year period.
- **Calculate the Average Yield**: Determine the average yield of the portfolio based on the yield curve.

A laddered portfolio provides a systematic approach for fixed-income investors to balance risk and reward.



## 1. Understanding the Laddering Strategy

In a laddered bond portfolio, bonds are purchased with staggered maturities, so that a portion of the portfolio matures at regular intervals. This strategy provides two main benefits:

- **Liquidity**: The portfolio always has bonds maturing periodically, offering liquidity to meet cash flow needs.
- **Interest Rate Risk Management**: As bonds mature, the proceeds can be reinvested at the current interest rates, reducing the impact of rate fluctuations over time.

This approach is especially useful in uncertain interest rate environments.



## 2. Hypothetical Yield Curve and Portfolio Setup

Let's set up a hypothetical yield curve and create a laddered portfolio with bonds maturing annually over a 10-year period.

Assume we have bonds with the following maturities and yields:


```python

import numpy as np
import matplotlib.pyplot as plt

# Define maturities and hypothetical yields for a 10-year laddered portfolio
maturities = np.arange(1, 11)  # 1 to 10 years
yields = np.array([0.015, 0.017, 0.019, 0.021, 0.023, 0.025, 0.027, 0.029, 0.03, 0.032])  # Hypothetical yields

# Plot the yield curve
plt.plot(maturities, yields, marker='o')
plt.xlabel("Maturity (Years)")
plt.ylabel("Yield")
plt.title("Hypothetical Upward Sloping Yield Curve")
plt.grid(True)
plt.show()
```


## 3. Laddered Portfolio Yield Calculation

In a laddered portfolio, we invest equally across each maturity. Let's calculate the portfolio's average yield based on our hypothetical yield curve.


```python

# Assume equal investment in each bond for simplicity
investment_per_bond = 1000  # Assume $1,000 per bond
total_investment = investment_per_bond * len(maturities)

# Calculate total yield based on laddered investments
portfolio_yield = np.sum(yields * investment_per_bond) / total_investment

print(f"Total Investment: ${total_investment}")
print(f"Average Yield of Laddered Portfolio: {portfolio_yield:.2%}")
```


## 4. Benefits and Risks of Laddering

A laddered bond portfolio provides:

- **Diversification of Maturities**: Reducing sensitivity to interest rate changes by spreading investments across a range of maturities.
- **Consistent Cash Flow**: Bonds maturing at regular intervals provide predictable cash flow for reinvestment or liquidity needs.

### Risks
- **Reinvestment Risk**: When bonds mature in a lower interest rate environment, reinvestment may yield lower returns.
- **Inflation Risk**: Laddering doesn't protect against inflation, which can erode purchasing power over time.

Laddered portfolios are effective in balancing return potential and risk, making them popular among fixed-income investors.
