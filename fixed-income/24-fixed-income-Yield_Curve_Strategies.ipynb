
# Yield Curve Strategies: Trading Based on Yield Curve Shifts

## Overview
Yield curve strategies involve trading bonds based on expected changes in the shape of the yield curve, which represents the relationship between bond yields and maturities. Common strategies include steepening, flattening, and butterfly spread approaches.

In this notebook, we will:

- **Understand Yield Curve Strategies**: Explore the basics of steepening, flattening, and butterfly spread strategies.
- **Use Real Data for Analysis**: Download U.S. Treasury yield data to analyze and simulate these strategies.
- **Example Calculations**: Apply each strategy using hypothetical shifts in the yield curve.

Yield curve strategies provide opportunities to profit from changes in economic conditions and interest rate expectations.



## 1. Basics of Yield Curve Strategies

There are several popular yield curve strategies:

- **Steepening Strategy**: Profit from an increase in the spread between short- and long-term yields, often expecting economic expansion or rising inflation.
- **Flattening Strategy**: Gain from a decrease in the spread between short- and long-term yields, usually anticipating slower economic growth or rate hikes.
- **Butterfly Spread**: Combines short, medium, and long-term bonds to profit from changes in the curvature of the yield curve.

Each strategy is based on expectations of interest rate movements and their impact on different maturities.



## 2. Downloading U.S. Treasury Yield Data

We'll use `yfinance` to download historical data for U.S. Treasury yields at different maturities. This data will help us analyze yield curve shapes and simulate the effects of steepening, flattening, and butterfly spread strategies.


```python

import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

# Download U.S. Treasury yield data for different maturities (e.g., 2-year, 10-year, and 30-year yields)
yields = {
    "2y": "^IRX",    # 2-year yield approximation
    "10y": "^TNX",   # 10-year yield
    "30y": "^TYX"    # 30-year yield
}

# Download historical data
data_2y = yf.download(yields["2y"], start="2010-01-01", end="2023-01-01")['Adj Close'] / 100  # Convert to decimal
data_10y = yf.download(yields["10y"], start="2010-01-01", end="2023-01-01")['Adj Close'] / 100
data_30y = yf.download(yields["30y"], start="2010-01-01", end="2023-01-01")['Adj Close'] / 100

# Plot yield data
plt.plot(data_2y.index, data_2y, label="2-Year Yield")
plt.plot(data_10y.index, data_10y, label="10-Year Yield")
plt.plot(data_30y.index, data_30y, label="30-Year Yield")
plt.xlabel("Date")
plt.ylabel("Yield")
plt.title("U.S. Treasury Yields (2y, 10y, 30y)")
plt.legend()
plt.grid(True)
plt.show()
```


## 3. Steepening Strategy

In a steepening strategy, we expect the yield curve to steepen, meaning the difference between short-term and long-term yields increases. Investors may go long on long-term bonds while shorting short-term bonds to profit from this shift.

### Example Calculation
Let's calculate the spread between the 2-year and 30-year yields to simulate a steepening trade.


```python
aligned_data = pd.concat([data_2y, data_30y], axis=1, join="inner")
aligned_data.columns = ["2y", "30y"]

spread_2y_30y = aligned_data["30y"] - aligned_data["2y"]

plt.plot(spread_2y_30y.index, spread_2y_30y, label="2-Year to 30-Year Spread")
plt.xlabel("Date")
plt.ylabel("Yield Spread")
plt.title("2-Year to 30-Year Yield Spread (Steepening Strategy)")
plt.ylim(-0.02, 0.05) 
plt.legend()
plt.grid(True)
plt.show()
```


## 4. Flattening Strategy

In a flattening strategy, we expect the yield curve to flatten, where the difference between short-term and long-term yields decreases. Investors may short long-term bonds and go long on short-term bonds.

### Example Calculation
We'll examine a hypothetical flattening by analyzing the spread between 2-year and 10-year yields.


```python
# Align data for the 2-year and 10-year yields to ensure dates match
aligned_data_2y_10y = pd.concat([data_2y, data_10y], axis=1, join="inner")
aligned_data_2y_10y.columns = ["2y", "10y"]

# Calculate the 2-year to 10-year yield spread
spread_2y_10y = aligned_data_2y_10y["10y"] - aligned_data_2y_10y["2y"]

# Plot the spread with a y-axis limit for better visualization
plt.plot(spread_2y_10y.index, spread_2y_10y, label="2-Year to 10-Year Spread")
plt.xlabel("Date")
plt.ylabel("Yield Spread")
plt.title("2-Year to 10-Year Yield Spread (Flattening Strategy)")
plt.ylim(-0.01, 0.03)  # Adjusted y-axis limit for better view
plt.legend()
plt.grid(True)
plt.show()
```


## 5. Butterfly Spread Strategy

A butterfly spread strategy involves combining short, medium, and long-term bonds to profit from changes in the yield curve's curvature. The strategy typically uses a short position in the medium-term bond while holding positions in short and long-term bonds.

### Example Calculation
We'll calculate the butterfly spread between 2-year, 10-year, and 30-year yields as:

\[ \text{Butterfly Spread} = \text{30y yield} - 2 \times (\text{10y yield}) + \text{2y yield} \]


```python
# Align data for the 2-year, 10-year, and 30-year yields to ensure dates match
aligned_data_butterfly = pd.concat([data_2y, data_10y, data_30y], axis=1, join="inner")
aligned_data_butterfly.columns = ["2y", "10y", "30y"]

butterfly_spread = aligned_data_butterfly["30y"] - 2 * aligned_data_butterfly["10y"] + aligned_data_butterfly["2y"]

plt.plot(butterfly_spread.index, butterfly_spread, label="Butterfly Spread (2y, 10y, 30y)")
plt.xlabel("Date")
plt.ylabel("Butterfly Spread")
plt.title("Butterfly Spread Strategy (Yield Curve)")
plt.ylim(-0.01, 0.01) 
plt.legend()
plt.grid(True)
plt.show()
```
