
# Relative Value Trading: Arbitraging Mispricing in Fixed Income Securities

## Overview
Relative value trading in fixed income involves identifying and exploiting mispricings between similar bonds or other fixed income securities. The goal is to capture profits from the spread between mispriced assets, assuming that the spread will converge over time.

In this notebook, we will:

- **Understand Relative Value Trading**: Learn the basics of relative value trading and how it applies to fixed income markets.
- **Use Real Data for Analysis**: Download and compare similar fixed income securities to detect mispricing opportunities.
- **Example Calculation**: Simulate a relative value trade by analyzing the spread between two similar bond ETFs.

Relative value trading is popular among institutional investors who seek to capitalize on small price discrepancies in bond markets.



## 1. Basics of Relative Value Trading

Relative value trading involves finding two securities that are fundamentally similar but temporarily mispriced relative to each other. Some common relative value trading pairs include:

- **Government Bonds of Similar Maturities**: For example, comparing U.S. Treasury bonds with similar maturity dates but slightly different yields.
- **Corporate Bonds in the Same Sector**: Analyzing bonds issued by companies in the same industry or with similar credit ratings.

The strategy typically involves going long on the undervalued bond and shorting the overvalued bond, with the expectation that the price spread will converge.



## 2. Downloading Data for Similar Fixed Income Securities

Weâ€™ll use `yfinance` to download data for two similar bond ETFs, such as iShares iBoxx $ Investment Grade Corporate Bond ETF (LQD) and Vanguard Total Corporate Bond ETF (VTC), and examine potential mispricing.

### Download and Compare Yields


```python

import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

# Download historical price data for two corporate bond ETFs (LQD and VTC)
etf1 = yf.download("LQD", start="2010-01-01", end="2023-01-01")
etf2 = yf.download("VTC", start="2010-01-01", end="2023-01-01")

# Calculate daily returns as a proxy for yield
yield_etf1 = etf1['Adj Close'].pct_change().dropna()
yield_etf2 = etf2['Adj Close'].pct_change().dropna()

# Align data to ensure consistent dates
aligned_yields = pd.concat([yield_etf1, yield_etf2], axis=1, join="inner")
aligned_yields.columns = ["ETF1 Yield", "ETF2 Yield"]

# Calculate the yield spread between the two ETFs
relative_value_spread = aligned_yields["ETF1 Yield"] - aligned_yields["ETF2 Yield"]

# Plot the relative value spread
plt.plot(relative_value_spread.index, relative_value_spread, label="Relative Value Spread (LQD - VTC)")
plt.xlabel("Date")
plt.ylabel("Yield Spread")
plt.title("Relative Value Spread between Corporate Bond ETFs (LQD and VTC)")
plt.legend()
plt.grid(True)
plt.show()
```


## 3. Example Relative Value Trade

In a relative value trade, we can exploit this yield spread by:

- **Going Long**: Buying the ETF with the lower yield (assuming it is undervalued).
- **Going Short**: Shorting the ETF with the higher yield (assuming it is overvalued).

The trade expects that the spread will revert to its historical average, allowing for a profitable exit.
