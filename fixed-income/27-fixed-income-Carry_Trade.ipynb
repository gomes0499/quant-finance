
# Carry Trade: Borrowing at Low Rates to Invest in Higher-Yield Instruments

## Overview
Carry trade is an investment strategy that involves borrowing funds at a lower interest rate to invest in higher-yield fixed income instruments. The goal is to profit from the difference between the borrowing and investment rates, also known as the "carry."

In this notebook, we will:

- **Understand the Carry Trade**: Explore the basics of carry trade and how it works in fixed income.
- **Analyze Real-World Interest Rates**: Use data to illustrate the potential profits and risks of carry trade strategies.
- **Example Calculation**: Calculate the potential carry for a hypothetical carry trade.

Carry trades are popular in fixed income markets but come with risks, including currency risk, interest rate risk, and leverage risk.



## 1. Basics of Carry Trade

In a carry trade, investors borrow funds at a low interest rate (often from a stable currency) and invest in assets with higher yields (such as emerging market bonds or corporate bonds). The profit comes from the spread between the borrowing cost and the investment yield.

Key risks include:

- **Currency Risk**: If the investment and borrowing are in different currencies, currency fluctuations can impact profits.
- **Interest Rate Risk**: Rising borrowing costs can reduce or eliminate the profit margin.
- **Leverage Risk**: Carry trades often involve leverage, amplifying potential gains and losses.



## 2. Example Carry Trade with Real Data

Let's simulate a carry trade where we borrow funds at a low U.S. Treasury yield and invest in a high-yield corporate bond ETF. Weâ€™ll use `yfinance` to download data on a U.S. Treasury bond ETF (e.g., SHY for 1-3 year Treasuries) and a high-yield corporate bond ETF (e.g., HYG).

### Download and Calculate Carry


```python

import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

# Download historical price data for U.S. Treasury bond ETF (SHY) and high-yield corporate bond ETF (HYG)
treasury_etf = yf.download("SHY", start="2010-01-01", end="2023-01-01")
high_yield_etf = yf.download("HYG", start="2010-01-01", end="2023-01-01")

# Calculate daily returns as a proxy for yield
yield_treasury = treasury_etf['Adj Close'].pct_change().dropna()
yield_high_yield = high_yield_etf['Adj Close'].pct_change().dropna()

# Align data to ensure consistent dates
aligned_yields = pd.concat([yield_treasury, yield_high_yield], axis=1, join="inner")
aligned_yields.columns = ["Treasury Yield", "High-Yield Corporate Yield"]

# Calculate carry (difference in yields)
carry = aligned_yields["High-Yield Corporate Yield"] - aligned_yields["Treasury Yield"]

# Plot the carry
plt.plot(carry.index, carry, label="Carry (High-Yield - Treasury)")
plt.xlabel("Date")
plt.ylabel("Carry (Yield Spread)")
plt.title("Carry Trade: Yield Spread between High-Yield Corporate and Treasury Bonds")
plt.legend()
plt.grid(True)
plt.show()
```


## 3. Example Carry Calculation

Assume a hypothetical carry trade where:

- **Borrowing Rate**: 1% (e.g., from Treasury bond ETF SHY)
- **Investment Yield**: 5% (e.g., high-yield corporate bond ETF HYG)

The expected carry would be the difference between the investment yield and borrowing rate.


```python

# Hypothetical carry calculation
borrowing_rate = 0.01  # Borrowing rate (1%)
investment_yield = 0.05  # Investment yield (5%)

# Calculate expected carry
expected_carry = investment_yield - borrowing_rate
print(f"Expected Carry: {expected_carry:.2%}")
```
