
# Roll-Down Return: Calculating Expected Returns from Holding Bonds

## Overview
The roll-down return, also known as the rolldown yield or rolldown effect, refers to the expected return achieved when a bond is held over a specified period and "rolls down" the yield curve. This return is influenced by the shape of the yield curve and is a strategy commonly used to enhance returns in a positively sloped yield curve environment.

In this notebook, we will:

- **Understand the Roll-Down Effect**: Explore how bond returns are influenced by changes in yield as they approach maturity.
- **Calculate Roll-Down Return**: Use hypothetical yield curve data to calculate expected returns over a given holding period.

This analysis helps fixed-income investors estimate returns beyond coupon income, based on the structure of the yield curve.



## 1. Understanding the Roll-Down Effect

The roll-down effect occurs when a bond is held as it moves closer to maturity, effectively shifting to a shorter maturity position on the yield curve. If the yield curve is upward sloping, the yield on the bond will decrease, increasing its price and generating an additional return beyond the bond's coupon.

The roll-down return can be an important factor for investors seeking to maximize returns in a fixed-income portfolio.



## 2. Hypothetical Yield Curve Data

We'll use a hypothetical yield curve to calculate the roll-down return for a bond with a specific initial maturity. The yield curve will be constructed for different maturities, and we'll calculate the expected roll-down return after a 1-year holding period.

For simplicity, assume the bond initially has 5 years to maturity.


```python

import numpy as np
import matplotlib.pyplot as plt

# Define maturities and hypothetical yields for an upward sloping yield curve
maturities = np.array([1, 2, 3, 4, 5, 7, 10])
yields = np.array([0.015, 0.017, 0.02, 0.022, 0.025, 0.027, 0.03])  # Hypothetical yields

# Plot the yield curve
plt.plot(maturities, yields, marker='o')
plt.xlabel("Maturity (Years)")
plt.ylabel("Yield")
plt.title("Hypothetical Upward Sloping Yield Curve")
plt.grid(True)
plt.show()
```


## 3. Roll-Down Return Calculation

Let's calculate the roll-down return for a bond with an initial maturity of 5 years, assuming a 1-year holding period.

### Calculation:
1. Find the initial yield at 5 years to maturity.
2. After holding the bond for 1 year, its effective maturity will be 4 years.
3. Calculate the price change due to the shift from the 5-year to the 4-year yield.


```python

from scipy.interpolate import interp1d

# Interpolate yield curve to find yields at specific maturities
yield_curve = interp1d(maturities, yields, kind='linear', fill_value="extrapolate")

# Initial parameters
initial_maturity = 5  # Initial maturity of the bond in years
holding_period = 1    # Holding period in years

# Yields at initial and adjusted maturities
yield_initial = yield_curve(initial_maturity)         # Yield at 5 years
yield_after_1_year = yield_curve(initial_maturity - holding_period)  # Yield at 4 years

# Calculate roll-down return (approximation)
roll_down_return = (yield_initial - yield_after_1_year) * 100  # Convert to percentage for clarity

print(f"Initial Yield (5 years): {yield_initial:.2%}")
print(f"Yield after 1-year holding (4 years): {yield_after_1_year:.2%}")
print(f"Estimated Roll-Down Return over 1 year: {roll_down_return:.2f}%")
```
