
# Regression Analysis: Linear Regression, Multiple Regression, and Logistic Regression

## Overview
Regression analysis is a statistical method for estimating relationships between variables. Itâ€™s widely used for prediction and forecasting in fields such as finance, economics, biology, and engineering.

In this notebook, we will cover:

- **Linear Regression**: Models the relationship between a dependent variable and one independent variable.
- **Multiple Regression**: Extends linear regression to include multiple independent variables.
- **Logistic Regression**: Used for binary classification, modeling the probability of a categorical outcome.



## 1. Linear Regression

Linear regression models the relationship between a dependent variable \( y \) and a single independent variable \( x \) using a line:

\[ y = \beta_0 + \beta_1 x + \epsilon \]

where \( \beta_0 \) is the intercept, \( \beta_1 \) is the slope, and \( \epsilon \) represents random error.


```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate sample data
np.random.seed(0)
x = np.random.rand(100, 1) * 10  # Independent variable
y = 2.5 * x + np.random.normal(0, 2, (100, 1))  # Dependent variable with some noise

# Fit linear regression model
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)

# Plot data and regression line
plt.scatter(x, y, color="blue", label="Data")
plt.plot(x, y_pred, color="red", label="Regression Line")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression")
plt.legend()
plt.show()

# Display regression coefficients
print("Intercept:", model.intercept_[0])
print("Slope:", model.coef_[0][0])
```


## 2. Multiple Regression

Multiple regression extends linear regression to model the relationship between a dependent variable \( y \) and multiple independent variables:

\[ y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n + \epsilon \]

Let's perform multiple regression in Python.


```python
# Generate sample data for multiple regression
np.random.seed(0)
X = np.random.rand(100, 3) * 10  # Three independent variables
y = 3 * X[:, 0] + 2 * X[:, 1] - 1.5 * X[:, 2] + np.random.normal(0, 2, 100)  # Dependent variable with noise

# Fit multiple regression model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Display regression coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```


## 3. Logistic Regression

Logistic regression is used for binary classification. It models the probability that a given input belongs to a particular category:

\[ P(y=1|x) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x_1 + \dots + \beta_n x_n)}} \]

Let's perform logistic regression to classify data into two categories.


```python

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Generate sample data for logistic regression
np.random.seed(0)
X = np.random.rand(100, 2) * 10
y = (X[:, 0] + X[:, 1] > 10).astype(int)  # Binary outcome based on a linear condition

# Fit logistic regression model
log_reg_model = LogisticRegression()
log_reg_model.fit(X, y)
y_pred = log_reg_model.predict(X)

# Plot data
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color="blue", label="Class 0")
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color="red", label="Class 1")
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Logistic Regression - Binary Classification")
plt.legend()
plt.show()

# Display accuracy
accuracy = accuracy_score(y, y_pred)
print("Model Accuracy:", accuracy)
```
